```python
marks = 85

if marks >= 85: #condition 1
    print('HD')
    
elif 75 <= marks < 85: #conditions 2
    print('D')
    
elif 60 < marks < 75: #condition 3
    print('C')
    
else:
    print('fail')
```

```python
first=int(input("enter the first number ")) #input 1 from user
second=int(input("enter the secomd number ")) #input 2 from user 
```

```python
#conditional statements

if first>second: #
    maximum=first
    minimum=second
else:
    maximum=second
    minimum=first
print("maximum",maximum)
print("minimum",minimum)
print("\n")
```

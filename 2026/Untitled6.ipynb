```python
import pygame
import sys

# Initialize Pygame
pygame.init()

# Function to create a window
def create_window(width, height, title, color):
    screen = pygame.display.set_mode((width, height))
    pygame.display.set_caption(title)
    screen.fill(color)
    pygame.display.flip()
    return screen

# First window
screen = create_window(500, 400, "First Window", (0, 128, 255))

# Main loop for first window
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN and event.key == pygame.K_SPACE:
            # Close first window and open second
            screen = create_window(600, 300, "Second Window", (255, 100, 0))

pygame.quit()
sys.exit()


```

```python
import pygame
import random
import sys

# Initialize PyGame
pygame.init()  #start game window loop

# Screen dimensions
WIDTH, HEIGHT = 800, 600 #
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Alien Invasion")

# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
GREEN = (0, 255, 0)

# Clock for controlling FPS
clock = pygame.time.Clock()

# Player settings
player_width, player_height = 50, 30
player_speed = 5
player = pygame.Rect(WIDTH // 2, HEIGHT - 50, player_width, player_height)

# Bullet settings
bullet_width, bullet_height = 5, 10
bullet_speed = 7
bullets = []

# Alien settings
alien_width, alien_height = 40, 30
alien_speed = 2
aliens = []
alien_spawn_delay = 40  # frames
alien_timer = 0

# Score
score = 0
font = pygame.font.SysFont(None, 36)

def draw_player():
    pygame.draw.rect(screen, GREEN, player)

def draw_bullets():
    for bullet in bullets:
        pygame.draw.rect(screen, RED, bullet)

def draw_aliens():
    for alien in aliens:
        pygame.draw.rect(screen, WHITE, alien)

def draw_score():
    score_text = font.render(f"Score: {score}", True, WHITE)
    screen.blit(score_text, (10, 10))

def game_over():
    over_text = font.render("GAME OVER", True, RED)
    screen.blit(over_text, (WIDTH // 2 - 80, HEIGHT // 2))
    pygame.display.flip()
    pygame.time.wait(2000)
    pygame.quit()
    sys.exit()

# Main game loop
while True:
    screen.fill(BLACK)

    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

    # Player movement
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and player.left > 0:
        player.x -= player_speed
    if keys[pygame.K_RIGHT] and player.right < WIDTH:
        player.x += player_speed
    if keys[pygame.K_SPACE]:
        if len(bullets) < 5:  # Limit bullets on screen
            bullets.append(pygame.Rect(player.centerx - bullet_width // 2, player.top - bullet_height, bullet_width, bullet_height))

    # Update bullets
    for bullet in bullets[:]:
        bullet.y -= bullet_speed
        if bullet.bottom < 0:
            bullets.remove(bullet)

    # Spawn aliens
    alien_timer += 1
    if alien_timer >= alien_spawn_delay:
        alien_x = random.randint(0, WIDTH - alien_width)
        aliens.append(pygame.Rect(alien_x, 0, alien_width, alien_height))
        alien_timer = 0

    # Update aliens
    for alien in aliens[:]:
        alien.y += alien_speed
        if alien.top > HEIGHT:
            game_over()

    # Collision detection
    for alien in aliens[:]:
        for bullet in bullets[:]:
            if alien.colliderect(bullet):
                aliens.remove(alien)
                bullets.remove(bullet)
                score += 10
                break

    # Draw everything
    draw_player()
    draw_bullets()
    draw_aliens()
    draw_score()

    pygame.display.flip()
    clock.tick(60)
```

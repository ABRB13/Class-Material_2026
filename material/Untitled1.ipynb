```python
#if/elif/else 

score = 65

if score >= 90:
    print("A")
elif 75 <= score < 90:
    print("B")
elif 50 <= score < 75:
    print("C")
else:
    print("fail")
```

```python
#while loops 

count = 1 

while count <= 3:   # i = 1 = true, "hello", count = 1 + 1, 2 # i = 2 = true, "hello", c = 2 + 1 = 3, i =3, true, "hello"
    print("hello") # count = 3 + 1 = 4. # i = 4, false, exit. 
    count = count + 1
```
